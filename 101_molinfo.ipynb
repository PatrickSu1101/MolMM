### Listing the details of the molecular descriptors

```python
from mordred import descriptors

# Loop through all descriptor groups in mordred.descriptors
for group_name in dir(descriptors):
    if(group_name[0].isupper()):
        group = getattr(descriptors, group_name)
        # print(dir(group))
        print('-'*40)
        print(group.__name__,group.__doc__,)
        # print(group)
        for i in dir(group):
            if(i[0].isupper()):
                print(i)
                # getattr(group,i).__name__)
                print(getattr(group,i).__doc__)
                print()
        # print(dir(group))
        # Only process descriptor groups (classes with __mordred_methods__)
        if hasattr(str(group), "__mordred_methods__"):
            print(f"\n=== {group_name} ===\n")
            for desc in group.__mordred_methods__:
                print(f"{desc.__name__}:\n{desc.__doc__}\n{'-'*40}")
```
